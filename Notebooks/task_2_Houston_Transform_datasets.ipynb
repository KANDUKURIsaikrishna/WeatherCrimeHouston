## Transform weather dataset

```python
import pandas as pd
data= pd.read_csv("/content/drive/MyDrive/cleaned_weather_df.csv")
```

```python
data.columns
```

```python
data=data.drop(["name","stations","description","icon"], axis=1)
data.head()
```

```python
data=data.rename(columns={"snow": "snow11"})
```

```python
data["conditions"].unique()  #'Partially cloudy', 'Clear', 'Rain,  Overcast', 'Snow, 'Overcast',
```

```python
data[["cond1","cond2","cond3"]] = data['conditions'].str.split(',', expand=True)
```

```python
data.columns
```

```python
data=pd.get_dummies(data, columns = ['cond1'])
data=pd.get_dummies(data, columns = ['cond2'])
data=pd.get_dummies(data, columns = ['cond3'])
```

```python
data.head()
```

```python
# Replace the underscore with a space
data.columns = [col.replace('cond1_', '') for col in data.columns]
data.columns = [col.replace('cond2_', '') for col in data.columns]
data.columns = [col.replace('cond3_', '') for col in data.columns]
data.columns = [col.replace(' ', '') for col in data.columns]
data
```

```python
data.columns
```

```python
data.shape
```

```python
#define function to merge columns with same names together
def same_merge(x): return ','.join(x[x.notnull()].astype(str))

#define new DataFrame that merges columns with same names together
df_new = data.groupby(level=0, axis=1).apply(lambda x: x.apply(same_merge, axis=1))
```

```python
df_new.shape
```

```python
df_new.columns
```

```python
df_new['cloudcover'].unique()
```

```python
df_new=df_new.replace('0,1', 1)
df_new
```

```python
df_new=df_new.drop(["conditions"], axis=1)
```

```python
df_new.to_csv("/content/drive/MyDrive/transformed_weather_data.csv")
```
