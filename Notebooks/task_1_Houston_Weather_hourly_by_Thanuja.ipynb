```python
houston_weather_df = pd.read_csv('data/weather/houston_weather.csv',skiprows=3);
```

```python
houston_weather_df.head(10)
```

```python
houston_weather_df['time_str'] = houston_weather_df['time'] + '-05:00'
houston_weather_df.head(10)
```

```python
houston_weather_df.dtypes

houston_weather_df['time'] = pd.to_datetime(houston_weather_df['time_str'], utc=True)
houston_weather_df.dtypes
```

```python
houston_weather_df = houston_weather_df.assign(
    occurrence_date=houston_weather_df["time"].dt.date, 
    occurrence_hour=houston_weather_df["time"].dt.hour
)
```

```python
houston_weather_df['is_day ()'].values
```

```python
fig = plt.figure(figsize = (10, 5))
w = 'occurrence_hour'
plt.title('nighttime hours')
sub_df=houston_weather_df[houston_weather_df['is_day ()'] ==0]
num_hours = houston_weather_df.size
night_hours=sub_df[[w, 'is_day ()']].groupby(w).count().reset_index().sort_values(by=w)
print(night_hours['is_day ()'].values)
x = night_hours[w].values.tolist()
y = (night_hours['is_day ()'].values / num_hours).tolist()
i = 0
while i < len(x) - 1 and x[i] + 1 == x[i+1]:
    i += 1
i += 1
if i < len(x):
    x = x[i:] + x[:i]
    y = y[i:] + y[:i]
x = [str(h) for h in x]
plt.plot(x, y)
    
plt.xlabel('hour')
plt.ylabel("crimes per hour")
        
plt.legend()
plt.show()
```
