```python
import pandas as pd
import os
import glob
```

```python
houston_crime_files = os.path.join("data/crime_fmt1", "*.xls")
```

```python
houston_crime_files_list= glob.glob(houston_crime_files)
print(houston_crime_files_list)
```

```python
crime_fmt1_df = pd.concat(map(pd.read_excel, houston_crime_files_list), ignore_index=True)
```

```python
#crime_fmt1_df.head(5)
crime_fmt1_df.dtypes

```

```python
crime_fmt1_df['street_name'] = crime_fmt1_df['Street Name'].combine_first(crime_fmt1_df['StreetName'])
crime_fmt1_df['block_range'] = crime_fmt1_df['Block Range'].combine_first(crime_fmt1_df['BlockRange'])
crime_fmt1_df['offense_count'] = crime_fmt1_df['# Of Offenses'].combine_first(crime_fmt1_df['# offenses']).combine_first(crime_fmt1_df['# Of']).combine_first(crime_fmt1_df['# Offenses'])
crime_fmt1_df.head(10)
```

```python
crime_fmt1_df.drop(columns=['Street Name', 'StreetName','Block Range','BlockRange','# Of Offenses','# offenses','# Of','# Offenses','Unnamed: 1','Field11'])
```

```python
crime_fmt1_df = crime_fmt1_df.rename(columns={'Date': 'occurrence_date', 'Hour': 'occurrence_hour', 'Offense Type':'offense_type','Beat':'beat','Premise':'premise','Suffix':'suffix','Type':'street_type'})
crime_fmt1_df.columns
```

```python
crime_fmt1_df = crime_fmt1_df[['occurrence_date', 'occurrence_hour', 'offense_type', 'offense_count', 'beat','premise','block_range','street_name','street_type','suffix']]
crime_fmt1_df.head(5)
crime_fmt1_df.dtypes
occurrence_hour_na_count_fmt1= crime_fmt1_df['occurrence_hour'].isna().sum()

print(occurrence_hour_na_count_fmt1)
crime_fmt1_df.tail(5)
crime_fmt1_df.head(5)
```

```python
# https://stackoverflow.com/questions/28669482/appending-pandas-dataframes-generated-in-a-for-loop
path = 'data/crime_fmt2/'
excel_files = glob.glob(os.path.join(path, "*.xlsx"))
crime_fmt2 = []
for f in excel_files:
    df = pd.read_excel(f,sheet_name=1,header=None,names=["occurrence_date","occurrence_hour","offense_type","offense_count","beat","premise","block_range","street_name","street_type","suffix"])
    print(f)
    print(df.shape)
    crime_fmt2.append(df)
                                   
crime_fmt2_df = pd.concat(crime_fmt2)
crime_fmt2_df.shape
crime_fmt2_df.columns
occurrence_hour_na_count_fmt2= crime_fmt2_df['occurrence_hour'].isna().sum()

print(occurrence_hour_na_count_fmt2)
crime_fmt2_df.tail(5)
#crime_fmt2_df.head(5)
```

```python
crime_fmt2_df.tail(5)
```

```python
path = 'data/crime_fmt3/'
excel_files = glob.glob(os.path.join(path, "*.xlsx"))
crime_fmt3 = []
for f in excel_files:
    df = pd.read_excel(f)
    print(f)
    print(df.shape)
    crime_fmt3.append(df)
                                   
crime_fmt3_df = pd.concat(crime_fmt3)
crime_fmt3_df.shape

crime_fmt3_df.head(5)
```

```python
print(crime_fmt3_df.columns)
crime_fmt3_df['occurrence_date'] = crime_fmt3_df['Occurrence\nDate'].combine_first(crime_fmt3_df['RMSOccurrenceDate'])
crime_fmt3_df['occurrence_hour'] = crime_fmt3_df['Occurrence\nHour'].combine_first(crime_fmt3_df['RMSOccurrenceHour'])
crime_fmt3_df['offense_count'] = crime_fmt3_df['Offense\nCount'].combine_first(crime_fmt3_df['OffenseCount'])
crime_fmt3_df['street_type'] = crime_fmt3_df['StreetType'].combine_first(crime_fmt3_df['Street\nType'])
crime_fmt3_df.head(10)
```

```python
crime_fmt3_df = crime_fmt3_df.drop(columns=['Incident', 'StreetNo','Occurrence\nDate', 'Occurrence\nHour', 'NIBRS\nClass','NIBRSClass','Offense\nCount','OffenseCount','RMSOccurrenceDate','City','RMSOccurrenceHour','ZIP Code','ZIPCode','MapLongitude', 'MapLatitude','Street\nType','StreetType'])
```

```python
crime_fmt3_df.head(5)
crime_fmt3_df.columns
```

```python
crime_fmt3_df = crime_fmt3_df.rename(columns={'NIBRSDescription': 'offense_type', 'Beat': 'beat', 'Premise':'premise','Block Range':'block_range','StreetName':'street_name','Suffix':'suffix'})
crime_fmt3_df.columns
```

```python
crime_fmt3_df = crime_fmt3_df[['occurrence_date', 'occurrence_hour', 'offense_type', 'offense_count', 'beat','premise','block_range','street_name','street_type','suffix']]
crime_fmt3_df.head(5)
occurrence_hour_na_count_fmt3 = crime_fmt3_df['occurrence_hour'].isna().sum()

print(occurrence_hour_na_count_fmt3)
crime_fmt3_df.tail(5)
crime_fmt3_df.head(5)
```

```python
crimes_df_list = [crime_fmt1_df, crime_fmt2_df,crime_fmt3_df]
houston_crime_df = pd.concat(crimes_df_list)
print(houston_crime_df.shape)
houston_crime_df.dropna(subset=['occurrence_hour', 'occurrence_date'], inplace=True)
# houston_crime_df['occurrence_hour'] = pd.to_numeric(houston_crime_df['occurrence_hour'], errors='coerce').astype(int)
print(houston_crime_df.shape)
print(houston_crime_df.columns)
print(houston_crime_df.tail(5))
houston_crime_df.dtypes
#houston_crime_df['occurrence_hour'] = houston_crime_df['occurrence_hour'].str.replace("'",'')
#houston_crime_df['occurrence_hour'] = houston_crime_df['occurrence_hour'].str.replace(r"\'", '')
houston_crime_df['occurrence_hour'] = houston_crime_df['occurrence_hour'].replace(regex=True,inplace=False,to_replace=r'\'',value=r'')
occurrence_hour_na_count = houston_crime_df['occurrence_hour'].isna().sum()
print(houston_crime_df.tail(5))
print(occurrence_hour_na_count)
#houston_crime_df['occurrence_hour'].fillna(-1, inplace=True)
houston_crime_df['occurrence_hour'] = houston_crime_df['occurrence_hour'].astype('int')
houston_crime_df['occurrence_date'] = houston_crime_df['occurrence_date'].astype('object')
```

```python
houston_crime_df['occurrence_date'] = pd.to_datetime(houston_crime_df['occurrence_date'],
                                                     format='%Y-%m-%d', errors='coerce', utc=True)
houston_crime_df.head(10)
```

```python
houston_weather_df = pd.read_csv('data/weather/houston_weather.csv',skiprows=3);
houston_weather_df.head(10)
```

```python
houston_weather_df.dtypes
houston_weather_df['time_str'] = houston_weather_df['time'] + '-05:00'
houston_weather_df['time'] = pd.to_datetime(houston_weather_df['time_str'], utc=True)
houston_weather_df.dtypes
```

```python
houston_weather_df = houston_weather_df.assign(
    occurrence_date=houston_weather_df["time"].dt.date, 
    occurrence_hour=houston_weather_df["time"].dt.hour
)
```

```python
houston_weather_df['occurrence_date'] = pd.to_datetime(houston_weather_df['occurrence_date'],
                                                       format='%Y-%m-%d', errors='coerce', utc=True)
```

```python
houston_weather_df.info()
```

```python
houston_weather_df.head(5)
houston_weather_df = houston_weather_df.drop(['time'], axis=1)
houston_weather_df.columns
```

```python
houston_weather_df = houston_weather_df[['occurrence_date', 'occurrence_hour', 'temperature_2m (°C)', 'relativehumidity_2m (%)',
       'dewpoint_2m (°C)', 'apparent_temperature (°C)', 'precipitation (mm)',
       'rain (mm)', 'snowfall (cm)', 'weathercode (wmo code)',
       'surface_pressure (hPa)', 'cloudcover (%)', 'cloudcover_low (%)',
       'cloudcover_mid (%)', 'cloudcover_high (%)',
       'et0_fao_evapotranspiration (mm)', 'vapor_pressure_deficit (kPa)',
       'windspeed_10m (km/h)', 'windgusts_10m (km/h)', 'is_day ()',
       'direct_radiation (W/m²)', 'diffuse_radiation (W/m²)']]
houston_weather_df.columns
```
