## Load Data

```python
import pandas as pd
data= pd.read_csv("/content/drive/MyDrive/transformed_weather_data.csv", index_col=None)
```

## Univariate Analysis

```python
# Size of the dataset
print("NB Rows:", data.shape[0])
print("NB Columns:", data.shape[1])
```

## Descriptive Statistics

```python
pd.set_option('display.max_columns', None)
DF=data.describe(include = 'all')
DF.loc['dtype'] = data.dtypes
DF.loc['nb null'] = data.isnull().sum()
DF
```

## Time series analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# Plot one example
fig, ax = plt.subplots(figsize=(15,5))
line = sns.lineplot(x='datetime',
                    y='humidity',
                    data=data)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Plot one example
fig, ax = plt.subplots(figsize=(15,5))
line = sns.lineplot(x='datetime',
                    y='temp',
                    data=data)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
